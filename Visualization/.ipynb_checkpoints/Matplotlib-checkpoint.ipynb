```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams["savefig.dpi"]
```

```python
x = np.linspace(0, 10, 1000)
y = np.sin(x)
z = np.cos(x**2)
```

```python
#figure
#plotmatplotlib
#figure
#
plt.figure(figsize=(8,4)) #figsizedpi80
plt.plot(x,y,label="$sin(x)$",color="red",linewidth=2)
#label : (legend)"$"matplotliblatex
#color : 
#linewidth : 
plt.plot(x,z,"b--",label="$cos(x^2)$")
plt.xlabel("Time(s)")  #X
plt.ylabel("Volt")  # Y
plt.title("PyPlot First Example")  #
plt.ylim(-1.2,1.2)  # Y
plt.legend() #
plt.show()
```

```python
#pyplotplt
import matplotlib.pyplot as plt


#
fig=plt.figure()

#p1211211
p1=fig.add_subplot(211)
x=[1,2,3,4,5,6,7,8]
y=[2,1,3,5,2,6,12,7]
p1.plot(x,y)

#p2211212
p2=fig.add_subplot(212)
a=[1,2]
b=[2,4]
p2.scatter(a,b)

plt.show()
```

## 3D

```python
#pyplotplt
import matplotlib.pyplot as plt
import numpy as np

#3D
from mpl_toolkits.mplot3d import Axes3D

#
fig = plt.figure()

#13D
ax = fig.add_subplot(111, projection='3d')

#X,Y,Z
# X = [1, 1, 2, 2] 
# Y = [3, 4, 4, 3]
# Z = [1, 100, 1, 1]
X = np.random.randn(10)
Y = np.random.randn(10)
Z = np.random.randn(10)
#4
ax.plot_trisurf(X, Y, Z)
plt.show()
```
