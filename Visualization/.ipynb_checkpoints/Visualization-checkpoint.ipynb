## Kernel Density Estimation Plots - KDEPlots

```python
# Standard
import numpy as np
import pandas as pd 
from numpy.random import randn
from pandas import Series,DataFrame
# Stats
from scipy import stats
# Ploting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = randn(25)
```

```python
sns.rugplot(dataset)
plt.ylim(0,1)
```

```python
plt.hist(dataset,alpha=0.3)   # alpha 透明度
sns.rugplot(dataset)
```

```python
sns.rugplot(dataset)
x_min = dataset.min() - 2
x_max = dataset.max() + 2

x_axis = np.linspace(x_min,x_max,100)

bandwidth = ( (4*dataset.std()**5) / (3*len(dataset))) ** 0.2

kernel_list = []
for data_point in dataset:
    #Create a kernel for each point and append to the kernek list
    kernel = stats.norm(data_point,bandwidth).pdf(x_axis)
    kernel_list.append(kernel)
    
    #Scale for plotting
    kernel = kernel / kernel.max()
    kernel = kernel * 0.4
    
    plt.plot(x_axis,kernel,color='grey',alpha=0.5)
    
plt.ylim(0,1)
```

```python
sum_of_kde = np.sum(kernel_list,axis=0)
fig = plt.plot(x_axis,sum_of_kde,color = 'indianred')
sns.rugplot(dataset)
plt.yticks([])
plt.suptitle('Sum of the basis functions')
```

```python
sns.kdeplot(dataset)   
```

```python
sns.rugplot(dataset,color = 'black')
for bw in np.arange(0.5,2,0.25):
    sns.kdeplot(dataset,bw=bw,lw=1.8,label=bw)
    
```

```python
kernel_options = ['biw','cos','epa','gau','tri','triw']
for kern in kernel_options:
    sns.kdeplot(dataset,kernel=kern,label = kern,shade=True)  # shade 阴影
```

```python
sns.kdeplot(dataset,vertical=True)
```

```python
sns.kdeplot(dataset,cumulative=True)
```

```python
mean = [0,0]

cov = [[1,0],[0,100]]

dataset2 = np.random.multivariate_normal(mean,cov,1000)

df = DataFrame(dataset2,columns = ['X','Y'])
sns.kdeplot(df.X,df.Y)
```

```python
sns.kdeplot(df.X,df.Y,shade=True)
```

```python
sns.kdeplot(df.X,df.Y,bw = 1)
```

```python
sns.kdeplot(df.X,df.Y,bw = 'silverman')
```

```python
sns.jointplot('X','Y',df,kind = 'kde')
```

## Regression Plots

```python
tips = sns.load_dataset('tips')
```

```python
tips.sex.value_counts()
```

```python
tips.head()
```

```python
sns.lmplot('total_bill','tip',tips)
```

```python
sns.lmplot('total_bill','tip',tips,scatter_kws ={'marker':'o','color':'indianred'},
                                      line_kws = {'linewidth':2,'color':'blue'})
```

```python
sns.lmplot('total_bill','tip',tips,order=5,
           scatter_kws ={'marker':'o','color':'indianred'},
            line_kws = {'linewidth':2,'color':'blue'})
```

```python
sns.lmplot('total_bill','tip',tips,fit_reg=False)  # 只要散点 
```

```python
tips.head()
```

```python
tips['tip_pect'] = 100 *( tips['tip'] / tips['total_bill'])
```

```python
tips.head()
```

```python
sns.lmplot('size','tip_pect',tips)
```

```python
sns.lmplot('size','tip_pect',tips,x_jitter=.1)
```

```python
sns.lmplot('size','tip_pect',tips,x_estimator=np.mean)
```

```python
sns.lmplot('total_bill','tip_pect',tips,hue='sex',markers=['x','o'])
```

```python
sns.lmplot('total_bill','tip_pect',tips,hue='day')
```

```python
sns.lmplot('total_bill','tip_pect',tips,lowess=True,line_kws = {'color':'black'})
```

```python
sns.lmplot('total_bill','tip_pect',tips)
```

```python
sns.regplot('total_bill','tip_pect',tips)  # lower level function that lmplot built on top of it;lmplot的下层方法
```

```python
fig ,(axis1,axis2) = plt.subplots(1,2,sharey =True)
sns.regplot('total_bill','tip_pect',tips,ax = axis1)
sns.violinplot(x=tips['size'],y=tips['tip_pect'],data = tips,color='indianred',ax = axis2)
```

## HeatMaps And Clustered Matrices

```python
flight_df = sns.load_dataset('flights')
```

```python
flight_df.head()
```

```python
flight_df = flight_df.pivot('month','year','passengers')
```

```python
flight_df
```

```python
sns.heatmap(flight_df)
```

```python
sns.heatmap(flight_df,annot=True,fmt='d')
```

```python
sns.heatmap(flight_df,center = flight_df.loc['June',1955])
```

```python
f , (axis1,axis2) = plt.subplots(2,1)

yearly_flight = flight_df.sum()
years = pd.Series(yearly_flight.index.values)
years = pd.DataFrame(years)

flights = pd.Series(yearly_flight.values)
flights = pd.DataFrame(flights)

year_df = pd.concat((years,flights),axis=1) 
year_df.columns = ['Year','Flights']

sns.barplot('Year',y = 'Flights',data = year_df,ax= axis1)

sns.heatmap(flight_df,cmap='Blues',ax=axis2,cbar_kws={'orientation':'horizontal'})
```

```python
sns.clustermap(flight_df)
```

```python
sns.clustermap(flight_df,col_cluster=False)
```

```python
sns.clustermap(flight_df,standard_scale=1)
```

```python
sns.clustermap(flight_df,standard_scale=0)
```

```python
sns.clustermap(flight_df,z_score=1)
```
